```python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd
import skrf as rf
import numpy as np
import re
import RawDFgenerator
```

```python
caminho = r'C:\Users\caval_5gfl5hy\OneDrive\Área de Trabalho\Nova pasta (5)-20251008T174947Z-1-001\Nova pasta (5)'
df = RawDFgenerator.find_file(caminho)
```

```python
df_filt1064nm = df.query('frequencia_ghz == 281760.0')
df2 = df_filt1064nm.drop(['arquivo'], axis =1)

inicio = 70
fim = 200
passo = 5
valores = np.arange(inicio, fim + 1, passo)

print(f"Valores base: {valores}")
num_valores = len(valores)
print(f"Número de valores únicos: {num_valores}")
print(f"Tamanho total da matriz será: {num_valores} x {num_valores} = {num_valores**2}")

coluna1 = np.repeat(valores, repeats=num_valores)

coluna2 = np.tile(valores, reps=num_valores)

matriz = np.column_stack((coluna1, coluna2))

df3 = df2.sort_values(by='id_nanopilar')
df3['L_x'] = matriz[:, 0]
df3['L_y'] = matriz[:, 1]

df3
```

```python
colunas = ['id_nanopilar', 'frequencia_hz', 'frequencia_ghz', 'S13_real', 'S13_imag', 'L_x', 'L_y']
df_TE = df3[colunas].copy()
df_TE['fase'] = np.arctan2(df_TE['S13_imag'], df_TE['S13_real'])

df_TE['transmitancia'] = np.sqrt((df_TE['S13_imag'])**2 + (df_TE['S13_real'])**2)
df_pivot = df_TE.pivot(index='L_y', columns='L_x', values='fase')
df_pivot2 = df_TE.pivot(index='L_y', columns='L_x', values='transmitancia')

colunas2 = ['id_nanopilar', 'frequencia_hz', 'frequencia_ghz', 'S24_real', 'S24_imag', 'L_x', 'L_y']
df_TM = df3[colunas2].copy()
df_TM['fase'] = np.arctan2(df_TM['S24_imag'], df_TM['S24_real'])

df_TM['transmitancia'] = np.sqrt((df_TM['S24_imag'])**2 + (df_TM['S24_real'])**2)
df_pivot3 = df_TM.pivot(index='L_y', columns='L_x', values='transmitancia')
df_pivot4 = df_TM.pivot(index='L_y', columns='L_x', values='fase')
```

```python
plt.figure(figsize=(8, 6))
ax = sns.heatmap(
    df_pivot,
    annot=False,
    fmt=".2f",
    cmap='jet',
    vmin=-np.pi,
    vmax=np.pi,linewidths=0.5,
    linecolor='black',
    cbar_kws={'ticks': [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]}
)
colorbar = ax.collections[0].colorbar
colorbar.set_ticklabels(['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])

plt.title('Phase TE', fontsize=16)
plt.xlabel('L_x (nm)', fontsize=12)
plt.ylabel('L_y (nm)', fontsize=12)
plt.show()
```
